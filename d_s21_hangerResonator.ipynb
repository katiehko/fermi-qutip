```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from math import pi
```

```python
w_0 = 10**4

Q_c = 10 ** 4
Q_i = 10 ** 6
Q_L = (Q_c ** (-1) + Q_i ** (-1)) ** (-1)

dlist = np.linspace(-10,10,200)
```

```python
i = 0
S_21_list = []
S_21_mag = []
S_21_ang = []
for delta in dlist:
    S_21 = 1 - ((Q_L/Q_c)/(1+(2j*Q_L*((delta)/w_0))))
    S_21_list.append(S_21)
    S_21_mag.append(abs(S_21))
    S_21_ang.append(np.angle(S_21))
```

```python
"""fig, ax = plt.subplots(1,1,figsize=(10,6))
ax.plot(dlist, S_21_list)
plt.xlabel("Detuning")
plt.ylabel("S$_{2,1}$ in dB")
plt.title("S$_{2,1}$")"""

fig, ax = plt.subplots(1,1,figsize=(10,6))
ax.plot(dlist, S_21_mag)
plt.xlabel("Detuning")
plt.ylabel("|S$_{2,1}$| in dB")
plt.title("S$_{2,1}$ Magnitude")

fig, ax = plt.subplots(1,1,figsize=(10,6))
ax.plot(dlist, S_21_ang)
plt.xlabel("Detuning")
plt.ylabel("Angle(S$_{2,1}$) in dB")
plt.title("S$_{2,1}$ in Phase")
```
